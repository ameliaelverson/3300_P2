```python
def get_disCount(col,sex,data):
    sex_indicator = 1 if sex == "boys" else 2
    grouped = data.groupby("NEWFPLRATIO - Amelia")
    series = grouped.aggregate(lambda x : (x[(x[col] == 1) & (x["Sex"] == sex_indicator) ].count()))["ADDADHD"]
    series.name = "".join(col.split()) + "_"+ sex
    return series
```

```python
disabilites = ['ADDADHD','Autism Spectrum',"Learning Disability"]
disabilites_counted = [get_disCount(d,s,disExcel) for d in disabilites for s in ["boys","girls"]]
boys_count = disIncome.aggregate(lambda x : (x[(x["Sex"] ==1) ].count()))["Sex"]
boys_count.name = "boys"
girls_count = disIncome.aggregate(lambda x : (x[(x["Sex"] ==2) ].count()))["Sex"]
girls_count.name = "girls"
```

```python
to_combine = disabilites_counted + [boys_count,girls_count]
```

```python
final = pd.concat(to_combine, axis = 1)
string_index = ["pad","<0.5","0.5-0.99","1.0-1.49","1.5-1.99",
"2-2.49","2.50-2.99","3.00-3.49","3.50-3.99","4-4.49","4.50-4.99","5+"]
final = pd.concat([final,pd.Series(string_index, name = "real_index")], axis = 1)
final = final[1:].set_index("real_index")
final
```

```python
final.to_json("disabilites_by_sex.json",orient = "index")
```
